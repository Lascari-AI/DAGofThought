```python
import json
import glob
from pathlib import Path

# Load all JSON files from education_system directory
json_files = glob.glob("../data/education_system/*.json")

responses = []
for file in json_files:
    with open(file) as f:
        responses.append(json.load(f)['parameters'])
```

```python
# dict_keys(['reasoning_process', 'findings_summary', 'remaining_questions', 'is_conclusion_premature', 'reason_for_premature_conclusion'])
```

```python
def build_prompt_for_gpt(responses):
    out = "Based on the full summary provided with the summary of each step and questions and why each set round is premature and the final response thought process, please create a text document detailing a full summary of the process. "
    out += "\n\n # Responses\n\n"
    for i, response in enumerate(responses):
        out += f"Response {i}:\n"
        out += f"Findings Summary: {response['findings_summary']}\n"
        out += f"Remaining Questions: {response['remaining_questions']}\n"
        out += f"Is Conclusion Premature: {response['is_conclusion_premature']}\n"
        out += f"Reason for Premature Conclusion: {response['reason_for_premature_conclusion']}\n"
        out += "\n"
        
    out += "\n\n # Final Response Thought Process\n\n"
    out += f"Final Response Thought Process: {json.dumps(responses[-1]['reasoning_process'], indent=4)}\n"
    return out

print(build_prompt_for_gpt(responses))
```
